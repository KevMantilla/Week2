```python
import pandas as pd
import os

# Función para cargar libros en DataFrames
def load_books(directory):
    texts = []
    filenames = []

    for filename in os.listdir(directory):
        if filename.endswith(".txt"):
            with open(os.path.join(directory, filename), 'r') as file:
                content = file.read()
                texts.append(content)
                filenames.append(filename)

    df = pd.DataFrame({'Filename': filenames, 'Content': texts})
    return df

# Función para realizar búsqueda booleana en los libros
def boolean_search(df, query):
    # Transformar la consulta en una lista de términos
    terms = query.lower().split()

    # Aplicar la búsqueda booleana
    results = df[df['Content'].str.lower().str.contains(' '.join(terms))]
    return results

# Directorio donde se encuentran los libros en formato txt
directory = '/content/Data '

# Cargar libros en un DataFrame
books_df = load_books(directory)

# Interfaz de búsqueda booleana
search_query = input("Enter your boolean search query: ")
search_results = boolean_search(books_df, search_query)

# Mostrar resultados
if not search_results.empty:
    print(f"Found matches in the following books:")
    for idx, row in search_results.iterrows():
        print(f"Book: {row['Filename']}")
        print(row['Content'][:500])  # Mostrar parte del contenido (primeros 500 caracteres)
        print("---")
    print(f"Total books containing matches: {len(search_results)}")
else:
    print("No matches found in any book.")
```
